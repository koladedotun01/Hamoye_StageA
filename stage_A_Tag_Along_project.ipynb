```python
#What is the total number of unique countries in the dataset?

total_uniq = food_df['Area'].nunique()
print('The total unique countries in the food dataset is:', total_uniq)
```

```python
#Which of these elements had the highest sum in 2018?
#Hint- Select columns ‘Y2018’ and ‘Element’, Perform a groupby operation on ‘Element’ on the selected dataframe and answer the question.

highest_element = food_df.groupby('Element').sum()['Y2018'].idxmax()
print('The element with the highest sum in Y2018 is:', highest_element)
```

```python
#What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

#note: Perform a groupby sum aggregation on ‘Item’

food_df.groupby('Item').sum().loc[['Animal fats'],['Y2014', 'Y2017']]
```

```python
#What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

new_food_df = food_df.describe().loc[['mean', 'std'],['Y2015']]
limit_decimal = round(new_food_df, 3)       #use round() function to keep the output 3 decimal places
limit_decimal
```

```python
#Which year had the highest correlation with ‘Element Code’?

food_df.corr()[['Element Code']].sort_values(by='Element Code', ascending=False)
```

```python
#What is the total number and percentage of missing data in 2016 to 2 decimal places?

missing_data_Y2016 = food_df.isnull().sum().loc['Y2016']
percent_missing = round((missing_data_Y2016 * 100 / len(food_df)), 2)

print('The total number of missing data in 2016 = ', missing_data_Y2016)
print('The % of the missing data in 2016 = ', percent_missing,'%')
```

```python
#Which of these elements had the 3rd lowest sum in 2018?

#note- Select columns ‘Y2018’ and ‘Element’, Perform a groupby operation on ‘Element’ on the selected dataframe and answer the question.

nums = food_df[['Y2018','Element']].groupby('Element').sum()
print("Original list of numbers: ",nums)
```

```python
#What is the total number of the sum of Production in 2014?

#note- Perform a groupby operation on ‘Element’ and use the resulting Dataframe to answer the question

food_df.groupby('Element').sum().loc[['Production'],['Y2014']]
```

```python
#What year has the highest sum of Import Quantity?

Highest_Quanlity = food_df.groupby('Element').sum().loc[['Import Quantity'],['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
y = Highest_Quanlity.max()
print(y)
```
